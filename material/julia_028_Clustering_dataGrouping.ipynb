## K-means 模型示範

## 載入資料

````julia
iris = RDatasets.dataset("datasets", "iris")
first(iris, 6)
````

````julia
typeof(iris)
````

````julia
X = Matrix(iris[:, 1:2])'
````

## K-means 分群模型
- `kmeans(X, k, [...])` -> KmeansResult
    - K-means clustering of the d×n data matrix X (each column of X is a d-dimensional data point; i.e. total n data points) into k clusters.
    - e.g. X[1,:] is the height and X[2,:] is the weight of total size(X,2) people. kmeans(X,k) groups these people into k groups iteratively, with an initial set of k points as centers. [Example and Explain](https://medium.com/@chih.sheng.huang821/%E6%A9%9F%E5%99%A8%E5%AD%B8%E7%BF%92-%E9%9B%86%E7%BE%A4%E5%88%86%E6%9E%90-k-means-clustering-e608a7fe1b43)

````julia
k = 3
result = kmeans(X, k)
````

## 取得分群
- assignments(R::ClusteringResult) -> Vector{Int}
    - Get the vector of cluster indices for each point.

- assignments(R)[i] is the index of the cluster to which the i-th point is
  assigned.

````julia
clustervec = assignments(result)
println("typeof(clustervec) = $(typeof(clustervec)); size(clustervec) = $(size(clustervec,1)),$(size(clustervec,2))")
````

````julia
iris[!, :cluster] = string.(clustervec);
````

````julia
unique(iris[:,:cluster])
````

````julia
sum(isequal.(iris[:,:cluster],"1"))
````

````julia
sum(isequal.(iris[:,:cluster],"2"))
````

````julia
sum(isequal.(iris[:,:cluster],"3"))
````

## 取得群中心

````julia
result.centers# the final centers of each groups
````

## 視覺化分群結果

````julia
plot(iris, x="SepalLength", y="SepalWidth", color="cluster", Geom.point)
````

## K-medoids 模型示範

## 前處理
- `euclidean(x, y)` is equal to `sqrt(sum((x - y) .^ 2))`

````julia
i= 1
a = Vector{Float64}(iris[i, 1:4])
````

````julia
j = 1
b = Vector{Float64}(iris[j, 1:4])
````

````julia
n = nrow(iris)
D = zeros(n, n)
for i = 1:n
    a = Vector{Float64}(iris[i, 1:4])
    for j = 1:n
        b = Vector{Float64}(iris[j, 1:4])
        D[i, j] = euclidean(a, b)
    end
end
````

## K-medoids 模型

````julia
k = 3
result = kmedoids(D, k)
````

## 取得分群

````julia
iris[!, :cluster] = string.(assignments(result));
````

## 取得群中心

````julia
result.medoids
````

````julia
iris[result.medoids, :]
````

## 視覺化分群結果

````julia
plot(iris, x="SepalLength", y="SepalWidth", color="cluster", Geom.point)
````

## 階層式分群法
- hierarchical clustering

````julia
result = hclust(D, linkage=:single)
````

## 繪製樹狀圖（dendrogram）

呈現階層式分群法

````julia
using Plots, StatsPlots
Plots.plot(result)
````
