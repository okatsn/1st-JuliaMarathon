````julia
using RDatasets, Plots
````

## 圓餅圖

````julia
haireyecolor = dataset("datasets", "HairEyeColor")
haircolor = by(haireyecolor, :Hair, Freq=:Freq => sum)
````

````julia
labels = haircolor[!, :Hair]
counts = haircolor[!, :Freq]
p = pie(labels, counts; size=(250, 200))
````

````julia
savefig(p, "pie.png")
````

````julia
n = 5000
x = randn(n)
y = -0.5x + randn(n)
p = marginalhist(x, y, fc=:plasma, bins=40)
````

## 3D 繪圖

````julia
n = 500
x = range(-8., stop=8., length=n)
y = range(-2., stop=2., length=n)
````

````julia
f(x,y) = x*exp(-(x-round(Int, x))^2-y^2)
````

````julia
p = plot(x, y, f, st=:surface)
````

````julia
savefig(p, "3d_function.png")
````

## 3D 動畫
- @gif 
    - create an animated gif file (and display it when returned to an IJulia cell)
    - for simple, one-off animations that you want to view immediately
    
- @animate will return an Animation object for later processing 
    - for anything more complex. Constructing Animation objects can be done when you need full control of the life-cycle of the animation (usually unnecessary though).

````julia
n = 500
x = range(-8., stop=8., length=n)
y = range(-2., stop=2., length=n)
f(x,y) = x*exp(-(x-round(Int, x))^2-y^2)
@gif for i in range(0, stop=2π, length=100)
    p = plot(x, y, f, st=:surface)
    plot!(p, camera=(15*cos(i), 40))
end
````

## 動畫輸出

````julia
p = plot(1)
ani = @animate for i in range(0, stop=2π, length=100)
    p = plot(x, y, f, st=:surface)
    plot!(p, camera=(15*cos(i), 40))
end
gif(ani, "animate.gif", fps = 30)
````
