```julia
### A Pluto.jl notebook ###
# v0.11.8

dt = 10;
sigma=sqrt(dt);  # variance=sigma^2
mm=0; #mean of dW
totalTime = 100000;
steps=Int64(round(totalTime/dt));
Nw=steps+1;
dW = sigma*randn(1,steps) .+ mm
F_C = 1;
D = 1;
Y0 = 0;
F_ext_to_F_C_ratio = 0.9;
F_ext = F_C*F_ext_to_F_C_ratio;
frictiontype = "Coulomb";
traceT = collect(range(1,step = dt,length = Nw));
b=sqrt(2*D);
```

```julia
if frictiontype == "Coulomb"
    drift(v) = -F_C*sign(v) + F_ext;
elseif frictiontype == "viscous"
    drift(v) = -F_C*v + F_ext;
end

function SDE(traceT,drift,Y0)
    dts = diff(traceT);
    traceY = fill(NaN,size(traceT))
    traceY[1] = Y0;
    y = Y0;
    for i = 2:length(traceT)
        y = y + drift(y)*dts[i-1] + b*dW[i-1]
        traceY[i] = y;
    end
    return traceY
end
```

```julia
traceY = SDE(traceT,drift,Y0)
using Plots
p1 = plot(traceT,traceY,xlabel = "t",ylabel= "v(t)");
p2 = histogram(traceY,bins = 40,normalize=:pdf,xlabel = "v",ylabel= "v(t)");
# :pdf, :density, :probability or :none
plot(p1,p2,layout = (1,2),legend= false)
```
